```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://www.flipkart.com/search?q=iphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
Allow = requests.get(URL)
```

```python
print(Allow)
```

```python
soup = BeautifulSoup(Allow.text)
```

```python
print(soup)
```

```python
data = soup.select("._4rR01T")
print(data)
```

```python
Iphones = []
for item in data:
    Iphones.append(item.text)
print(Iphones)
```

```python
Mobiles = pd.DataFrame(Iphones,columns = ["Iphones"])
```

```python
Mobiles
```

```python
Price_data = soup.select("._1_WHN1")
```

```python
Price_data
```

```python
Mobiles["Deals"] = Price_data
```

```python
Mobiles
```

```python
Original_data = soup.select("._27UcVY")
```

```python
Original_data
```

```python
Mobiles[" price"] = Original_data
```
